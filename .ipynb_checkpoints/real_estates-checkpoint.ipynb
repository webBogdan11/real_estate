# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 
df = pd.read_csv('data_sets/real_estate_data.csv', sep = '\t')
df.info()
df.head(10)
```

### Вывод 

В данных очень много пропусков, на большинство пропусков подходящей замены будет найти трудно, по этому придется в некоторых местах оставить

## Предобработка данных

```python
# Проверка на дубликаты

print(df.duplicated().sum())

# Подсчет всех пропусков по столбцам
manual_columns = []

for i in df.columns:
    print(i, df[i].isna().sum())
    if df[i].isna().sum() > 0:
        manual_columns.append(i)

# Собрали все столбци которые были заданы вручную и обнаружены наны
print(manual_columns)

# Посчитали количество строк в которых большинство пропуски их оказалось 270
df_manual_nan = df
for i in manual_columns:
    if i == 'locality_name' or i== 'floors_total' or i== 'days_exposition':
        continue
    else:
        df_manual_nan = df_manual_nan[df_manual_nan[i].isna()]

print(df_manual_nan)
# Удаляем 270 строк где большинство пропусков
df.drop(labels = df_manual_nan.index ,axis = 0, inplace = True)
df.reset_index(drop = True, inplace = True)
df
```

Вывел количество NAN в каждом столбце

Решил удалить столбци в которых большинство данных не заполнено вручную, так как посчитал что это не правильные обьявления и они только портят данные

```python
#Проверка на дупликаты
print(len(df['locality_name'].unique()))
print(len(df['locality_name'].str.lower().unique()))

display(df[df['locality_name'].isna()].head(10))
print(df['locality_name'].isna().sum())
df['locality_name'].fillna(value = 'Не указан', inplace = True)
print(df['locality_name'].isna().sum())
```

Обработал значения столбца locality_name, так как данных не достаточно для точного определение города и нанов в этом столбце не много, то решил просто заменить их на строку Не указан

```python
#Заметил аномальные значния
print(df['ceiling_height'].unique())


# Посчитал 128 аномальных значений 
print(df.query('ceiling_height <= 2.0 or ceiling_height >= 4.0')['ceiling_height'].count())

# Меняю Наны на медиану и Медианой заполняю аномальные значения до 2 метров и больше 4 метров
ceiling_height_median = df['ceiling_height'].median()
df['ceiling_height'] = df['ceiling_height'].apply(lambda x: ceiling_height_median if x <= 2.0 or x >= 4.0 else x)
df['ceiling_height'] = df['ceiling_height'].fillna(ceiling_height_median)

# Проверка 
print(df['ceiling_height'].isna().sum())
```

При обработке данных столбца с высотой потолка определил, что много аномальных значений и их нужно заменить медианным, так как большинство значений высоты потолков не отличаются от медианы на большое числов и также этим значением заменил наны, считаю это оптимальным решением

```python
# Изучения данных столбца количества этажей + подсчет изначально пропущенных значений
print(df['floors_total'].unique())
print(df[df['floors_total'].isna()].head(50))
print(df['floors_total'].isna().sum())

# Обьявление функции для определение одинаковых по 3 параметрам: Название населенного пункта, Растояние до центра, растояние до аэропорта
def determine_house_for_total_floors(x):
    if pd.isnull(x['floors_total']):
        median_total_floors = df[(df['locality_name'] == x['locality_name']) 
                                & (df['cityCenters_nearest'] == x['cityCenters_nearest'])
                                & (df['airports_nearest'] == x['airports_nearest'])]['floors_total'].median()
        if median_total_floors < x['floor']:
            return x['floors_total']
        else:
            return median_total_floors
    else:
        return x['floors_total']

# Применение функции и замена нанов на количество этажей в одинаковых домах
df['floors_total'] = df.apply(func = determine_house_for_total_floors, axis = 1)
print(df['floors_total'].isna().sum())

# Остались пропуски так как может быть нету больше квартир с одинаковыми 3 параметрами что в функци либо для всех квартир в доме не указано количество этажей
```

Для замены нанов в количестве этажей пришлось вычислить похожие дома (одинаковые) по трем параметрам город, растояние до центра, и растояние до аиропорта, использовали эти значения как координаты и заполнили этим, но некторые дома не получилось так сопоставить по этому остались наны

```python
# Изученее данных колонки жилая площадь
print(df['living_area'].unique())

df['living_area'].value_counts().head(50)

df.query('living_area > total_area')

print(df['living_area'].isna().sum())

# Задание функции для нахождения строк с одинаковой площадью и комнат
def determine_house_for_area(x, type_of_area):
    if pd.isnull(x[type_of_area]):
        median_living_area = df[(df['locality_name'] == x['locality_name']) 
                                & (df['total_area'] == x['total_area'])
                                & (df['rooms'] == x['rooms'])][type_of_area].median()

        return median_living_area
    else:
        return x[type_of_area]

def determine_house_for_living_area(x):
    return determine_house_for_area(x, type_of_area = 'living_area')

# Вместо пропусков вставляем медианное значение для похожих строк
df['living_area'] = df.apply(func = determine_house_for_living_area, axis = 1)

print(df['living_area'].isna().sum())

# Удаляем странные данные где жилая площадь меньше 6
df.drop(labels = df.query('living_area < 6').index ,axis = 0, inplace = True)
df.reset_index(drop = True, inplace = True)
```


<div class="alert alert-success">

Для заполнение пропущенных значений здесь посчитали медианное значение среди квартир с одинаковой площадью и комнат

Для тех у кого все равно остались наны оставляем их

Удалил аномальные значения когда жилая плозаль меньше 6

```python
# Изучение данных
print(df['kitchen_area'].isna().sum())
display(df.query('kitchen_area > total_area'))

# Запуск функции для заполнений нанов
def determine_house_for__area(x):
    return determine_house_for_area(x, type_of_area = 'kitchen_area')

df['kitchen_area'] = df.apply(func = determine_house_for__area, axis = 1)
print(df['kitchen_area'].isna().sum())

# Удаление экстримально маленьких данных
df.drop(labels = df.query('kitchen_area < 3').index ,axis = 0, inplace = True)
df.reset_index(drop = True, inplace = True)
```

Самая высокая положительная корреляция между ценной за кв метр и полной стоимостью, между всей площадью и числом комнат

```python
most_popular_cities = df['locality_name'].value_counts().head(10).index

max_mean_value = 0
lowest_mean_value = float('inf') 

for i in most_popular_cities:
    mean_value = df.query('locality_name == @i')['square_meter_price'].mean()

    
    if max_mean_value < mean_value:
        max_mean_value = mean_value 
        city_max = i
        
    if lowest_mean_value > mean_value:
        lowest_mean_value = mean_value
        city_lowest = i    
    
    print(i, round(mean_value, 3))
print()
print('Самая большая цена на квадратный метр', round(max_mean_value, 3), city_max)
print('Самая маленькая цена на квадратный метр', round(lowest_mean_value,3), city_lowest)
```

```python
current_df = df.query('locality_name == "Санкт-Петербург"')

# Формирование столбца с километрами
current_df.loc[:, 'cityCenters_distance'] = df.loc[:,'cityCenters_nearest'].apply(func = lambda x: x if pd.isna(x) else int(x / 1000))
current_df.reset_index(drop= True, inplace = True)
current_df['cityCenters_distance'].hist(bins=50)
plt.show()

#Построение графика определения центрльной зоны
price_kilometrs = current_df.pivot_table(index = 'cityCenters_distance', values = 'last_price')
plt.plot(price_kilometrs.index, price_kilometrs['last_price']) 

# От до 10 километров это центральная зона
```
